# Data Visualization Chart Types

**1. Basic Charts**

| Chart Type  | Description | Matplotlib/Seaborn/Plotly Example |
|-------------|------------|----------------------------------|
| **Line Chart** | Shows trends over time. Used for continuous data. | `plt.plot()`, `sns.lineplot()`, `px.line()` |
| **Bar Chart** | Compares categorical data using bars. | `plt.bar()`, `sns.barplot()`, `px.bar()` |
| **Histogram** | Shows frequency distribution of numerical data. | `plt.hist()`, `sns.histplot()`, `px.histogram()` |
| **Pie Chart** | Represents proportions as slices of a circle. | `plt.pie()`, `px.pie()` |

**2. Distribution Charts**

| Chart Type  | Description | Matplotlib/Seaborn/Plotly Example |
|-------------|------------|----------------------------------|
| **Box Plot (Box-and-Whisker Plot)** | Shows data distribution, median, and outliers. | `plt.boxplot()`, `sns.boxplot()`, `px.box()` |
| **Violin Plot** | Like a box plot but shows KDE (distribution). | `sns.violinplot()`, `px.violin()` |
| **Density Plot (KDE Plot)** | Smooths histogram into a probability density function. | `sns.kdeplot()`, `px.density_contour()` |


**3. Scatter and Relationship Charts**

| Chart Type  | Description | Matplotlib/Seaborn/Plotly Example |
|-------------|------------|----------------------------------|
| **Scatter Plot** | Shows relationship between two numerical variables. | `plt.scatter()`, `sns.scatterplot()`, `px.scatter()` |
| **Bubble Chart** | A scatter plot where point size represents a third variable. | `px.scatter()` with `size` argument |
| **Pairplot (Scatter Matrix)** | Shows relationships between multiple numerical features. | `sns.pairplot()` |


**4. Advanced Charts**

| Chart Type  | Description | Matplotlib/Seaborn/Plotly Example |
|-------------|------------|----------------------------------|
| **Heatmap** | Shows correlations between multiple variables using color intensity. | `sns.heatmap()`, `px.imshow()` |
| **Treemap** | Nested rectangles to visualize hierarchical data. | `px.treemap()` |
| **Radar Chart (Spider Plot)** | Used for comparing multiple variables. | `plt.polar()`, `plotly.graph_objects.Figure()` |
| **Sunburst Chart** | Shows hierarchical data like a multi-level pie chart. | `px.sunburst()` |


**5. Time Series and Financial Charts**

| Chart Type  | Description | Matplotlib/Seaborn/Plotly Example |
|-------------|------------|----------------------------------|
| **Time Series Line Chart** | Shows changes over time. | `plt.plot()`, `px.line()` |
| **Candlestick Chart** | Used for stock market data (open-high-low-close). | `plotly.graph_objects.Candlestick()` |


**6. Specialized Charts**

| Chart Type  | Description | Matplotlib/Seaborn/Plotly Example |
|-------------|------------|----------------------------------|
| **Word Cloud** | Shows word frequency in text. | `wordcloud.WordCloud()` |
| **Network Graph** | Visualizes relationships (nodes and edges). | `networkx.draw()` |
| **Chord Diagram** | Shows flow between categories. | `plotly.graph_objects.Sankey()` |



## 1. Basic Charts

### Line Chart (Trends Over Time)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate Sample Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Set Theme & Color Palette
sns.set_theme(style="darkgrid")
colors = sns.color_palette("husl", 2)  # Vibrant Power BI-like colors

# Create Figure
plt.figure(figsize=(10, 5))

# Plot Lines
plt.plot(x, y1, label="Sine Wave", color=colors[0], linewidth=2, linestyle='-')
plt.plot(x, y2, label="Cosine Wave", color=colors[1], linewidth=2, linestyle='dashed')

# Titles & Labels
plt.title("Line Chart - Sine & Cosine Trends", fontsize=16, fontweight="bold")
plt.xlabel("Time", fontsize=12)
plt.ylabel("Value", fontsize=12)

# Grid & Legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)

# Show Plot
plt.show()
```

### Bar Chart (Category Comparison)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate Sample Data
categories = ["A", "B", "C", "D", "E"]
values = np.random.randint(10, 100, size=len(categories))

# Set Theme & Colors
sns.set_theme(style="whitegrid")
colors = sns.color_palette("coolwarm", len(categories))

# Create Figure
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, values, color=colors, edgecolor="black")

# Titles & Labels
plt.title("Bar Chart - Category Comparison", fontsize=16, fontweight="bold")
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Annotate Bars with Values
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2 - 0.15, bar.get_height() + 2,
             str(bar.get_height()), fontsize=12, fontweight="bold")

# Show Plot
plt.show()
```

### Histogram (Distribution of Data)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate Sample Data
data = np.random.normal(loc=50, scale=15, size=500)  # Mean=50, Std=15

# Set Theme & Colors
sns.set_theme(style="whitegrid")

# Create Figure
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=20, kde=True, color="royalblue", edgecolor="black", linewidth=1.2)

# Titles & Labels
plt.title("Histogram - Data Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Show Plot
plt.show()
```

### Pie Chart (Proportion Representation)

```python
import matplotlib.pyplot as plt
import numpy as np

# Modern Style
# plt.style.use('dark_background')

# Data
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
sizes = np.random.randint(5, 20, len(labels))  # Random percentages
colors = plt.cm.Greens(np.linspace(0.3, 1, len(labels)))  # Green color palette

# Plot
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes, labels=labels, autopct='%1.1f%%',
    startangle=90, colors=colors, pctdistance=0.85,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Style
for text in texts:
    text.set_color("black")
for autotext in autotexts:
    autotext.set_color("black")

ax.set_title("Elegant Pie Chart", fontsize=16, fontweight='bold')
plt.show()
```

## 2. Distribution Charts

### 1. Box Plot (Box-and-Whisker Plot)
- **Displays the median, quartiles, and outliers** in the dataset.
- Helps in identifying **skewness and spread** of the data.
- **Best For:** Comparing distributions across different categories.
- **Example:** Seaborn Box Plot, Plotly Interactive Box Plot.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data
np.random.seed(42)
data = pd.DataFrame({
    "Category": np.random.choice(["A", "B", "C", "D"], size=200),
    "Values": np.random.normal(loc=50, scale=15, size=200)
})

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Category", y="Values", data=data, palette="Set2", width=0.6, linewidth=2)
plt.title("Box Plot: Distribution of Values by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### 2. Violin Plot
- **Combines a Box Plot with a Kernel Density Estimate (KDE).**
- Shows the **distribution shape** and **probability density**.
- **Best For:** Understanding both summary statistics and data distribution.
- **Example:** Seaborn Violin Plot, Plotly Interactive Violin Plot.


```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="Category", y="Values", data=data, palette="coolwarm", inner="quartile", linewidth=2)
plt.title("Violin Plot: Value Distribution by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### 3. KDE Plot (Density Plot)
- A **smoothed histogram** that estimates the **probability density function** of data.
- **Best For:** Understanding data distribution without binning issues.
- **Example:** Seaborn KDE Plot, Plotly Density Contour.

```python
# KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data["Values"], shade=True, color="darkblue", linewidth=2)
plt.title("Kernel Density Estimate (KDE) Plot", fontsize=14)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(linestyle="--", alpha=0.7)
plt.show()
```

### 4. Box Plot & Violin Plot Combined
- **Useful for comparing Box Plot and Violin Plot together.**
- Highlights both **summary statistics** and **distribution shape**.
- **Example:** Seaborn Box + Violin Plot.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Box Plot
sns.boxplot(x="Category", y="Values", data=data, ax=axes[0], palette="Set2", width=0.6, linewidth=2)
axes[0].set_title("Box Plot")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Values")
axes[0].grid(axis="y", linestyle="--", alpha=0.7)

# Violin Plot
sns.violinplot(x="Category", y="Values", data=data, ax=axes[1], palette="coolwarm", inner="quartile", linewidth=2)
axes[1].set_title("Violin Plot")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Values")
axes[1].grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

### 5. Interactive Box Plot with Plotly
- **Plotly provides an interactive version** of Box Plot.
- Allows users to **hover, zoom, and explore outliers dynamically**.
- **Best For:** Exploratory Data Analysis (EDA).
- **Example:** Plotly Box Plot.

```python
import plotly.express as px

# Interactive Box Plot
fig = px.box(data, x="Category", y="Values", color="Category", title="Interactive Box Plot (Plotly)")
fig.show()
```

### 6. Interactive Violin Plot with Plotly
- **Interactive Violin Plot** to explore **density distributions dynamically**.
- Enables **zooming and filtering** for deeper insights.
- **Example:** Plotly Violin Plot.

```python
# Interactive Violin Plot
fig = px.violin(data, x="Category", y="Values", color="Category", box=True, points="all",
                title="Interactive Violin Plot (Plotly)")
fig.show()
```

### 3. Radar Chart (Spider Plot)
- **Purpose:** Used for comparing multiple variables across different categories.
- **Best For:** Performance evaluation, player stats, etc.

**Matplotlib Radar Chart**

```python
import numpy as np

# Sample Data
labels = ["Speed", "Strength", "Agility", "Endurance", "Intelligence"]
values = [8, 6, 7, 9, 8]

# Extend to close the circle
values += values[:1]
angles = np.linspace(0, 2 * np.pi, len(values))

# Matplotlib Radar Chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color="blue", alpha=0.3)
ax.plot(angles, values, color="blue", linewidth=2)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title("Radar Chart (Spider Plot)")
plt.show()
```

**Plotly Radar Chart**

```python
import plotly.graph_objects as go

# Plotly Radar Chart
fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=[8, 6, 7, 9, 8, 8], theta=labels + [labels[0]],
    fill="toself", name="Performance"
))
fig.update_layout(polar=dict(radialaxis=dict(visible=True)),
                  title="Interactive Radar Chart")
fig.show()
```

### 3. Chord Diagram
- **Purpose:** Shows relationships or flow between categories.
- **Best For:** Supply chains, migration data, flow of transactions.

```python
import plotly.graph_objects as go

# Define data for the Sankey diagram (Chord Diagram Equivalent)
data = {
    "source": [0, 0, 1, 2, 3, 3],  # Source indices (map to labels)
    "target": [1, 2, 3, 4, 5, 4],  # Target indices
    "value": [10, 20, 30, 40, 50, 60]  # Flow values
}

# Define labels for nodes
labels = ["A", "B", "C", "D", "E", "F"]

# Create Sankey (Chord Equivalent) Diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15, thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color=["blue", "green", "red", "orange", "purple", "pink"]
    ),
    link=dict(
        source=data["source"],
        target=data["target"],
        value=data["value"]
    )
))

fig.update_layout(title_text="Chord Diagram (Sankey) Visualization", font_size=12)
fig.show()
```

# Different Ways to Build Subplots in Matplotlib

Subplots are used to arrange multiple plots in a single figure. Here are various ways to create subplots:

## 1. Using plt.subplot() (Single Plot at a Time)
This method adds one subplot at a time.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1, y2, y3 = np.sin(x), np.cos(x), np.tan(x)

plt.figure(figsize=(10, 6))  # Set figure size

# First subplot (1st row, 3 columns, 1st position)
plt.subplot(1, 3, 1)
plt.plot(x, y1)
plt.title("Sine Wave")

# Second subplot (1st row, 3 columns, 2nd position)
plt.subplot(1, 3, 2)
plt.plot(x, y2)
plt.title("Cosine Wave")

# Third subplot (1st row, 3 columns, 3rd position)
plt.subplot(1, 3, 3)
plt.plot(x, y3)
plt.title("Tangent Wave")

plt.tight_layout()  # Adjust spacing
plt.show()
```

- Uses `plt.subplot(rows, cols, index)`
- Creates subplots sequentially
- Manually called for each subplot

## 2. Using plt.subplots() (Better for Multiple Plots)
This method creates all subplots at once and returns figure (`fig`) and axes (`ax`).

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))  # 1 row, 3 columns

ax[0].plot(x, y1)
ax[0].set_title("Sine Wave")

ax[1].plot(x, y2)
ax[1].set_title("Cosine Wave")

ax[2].plot(x, y3)
ax[2].set_title("Tangent Wave")

plt.show()
```

- More Pythonic and avoids repeated `plt.subplot()` calls
- Returns axes array (`ax`), making it easy to loop through subplots

## 3. Using plt.subplots() with Loop
Instead of manually assigning plots, use a loop.


```python
fig, ax = plt.subplots(2, 2, figsize=(10, 6))  # 2 rows, 2 columns

functions = [np.sin, np.cos, np.tan, np.exp]
titles = ["Sine", "Cosine", "Tangent", "Exponential"]

for i, axi in enumerate(ax.flatten()):  # Flattening 2D array
    axi.plot(x, functions[i](x))
    axi.set_title(titles[i])

plt.tight_layout()
plt.show()
```

- Avoids repetitive code
- Useful for large grids of plots

## 4. Using plt.GridSpec() for Unequal Subplot Sizes
When you need different-sized subplots.

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(3, 3)  # 3x3 grid

# Large subplot spanning multiple cells
ax1 = plt.subplot(gs[0, :])  # First row, spanning all columns
ax1.plot(x, y1)
ax1.set_title("Sine Wave (Large)")

# Two smaller subplots
ax2 = plt.subplot(gs[1, :2])  # Second row, first two columns
ax2.plot(x, y2)
ax2.set_title("Cosine Wave")

ax3 = plt.subplot(gs[1:, 2])  # Second and third row, last column
ax3.plot(x, y3)
ax3.set_title("Tangent Wave")

ax4 = plt.subplot(gs[2, :2])  # Last row, first two columns
ax4.plot(x, np.exp(x / 10))
ax4.set_title("Exponential Growth")

plt.tight_layout()
plt.show()
```

- Flexible layout customization
- Allows subplots to span multiple rows/columns

## 5. Using fig.add_subplot() (Object-Oriented Approach)
This method manually adds subplots to a figure.


```python
fig = plt.figure(figsize=(10, 5))

ax1 = fig.add_subplot(1, 2, 1)  # 1st subplot
ax1.plot(x, y1)
ax1.set_title("Sine Wave")

ax2 = fig.add_subplot(1, 2, 2)  # 2nd subplot
ax2.plot(x, y2)
ax2.set_title("Cosine Wave")

plt.show()
```

- Similar to `plt.subplot()` but object-oriented
- Explicitly associates subplots with figure

## 6. Using plt.Figure().add_axes() for Absolute Positioning
This method gives precise control over subplot positions.


```python
fig = plt.figure(figsize=(8, 5))

# Add a subplot at position (left, bottom, width, height)
ax1 = fig.add_axes([0.1, 0.5, 0.4, 0.4])  # Top left
ax2 = fig.add_axes([0.55, 0.5, 0.4, 0.4])  # Top right
ax3 = fig.add_axes([0.1, 0.1, 0.4, 0.4])  # Bottom left
ax4 = fig.add_axes([0.55, 0.1, 0.4, 0.4])  # Bottom right

ax1.plot(x, y1)
ax1.set_title("Sine")

ax2.plot(x, y2)
ax2.set_title("Cosine")

ax3.plot(x, y3)
ax3.set_title("Tangent")

ax4.plot(x, np.exp(x / 10))
ax4.set_title("Exponential")

plt.show()
```

- Gives absolute control over subplot placement
- Useful for complex layouts (e.g., inset plots)

## 7. Creating Subplots Dynamically (Interactive Dashboards)

```python
rows, cols = 2, 2
fig, axes = plt.subplots(rows, cols, figsize=(8, 6))

for row in range(rows):
    for col in range(cols):
        ax = axes[row, col]
        ax.plot(x, np.random.rand(len(x)))  # Random plots
        ax.set_title(f"Plot ({row+1}, {col+1})")

plt.tight_layout()
plt.show()
```

- Dynamically generates grids of subplots
- Useful for plotting datasets of unknown size

## 8. Creating Subplots with Different Y-Axes (Twin Axes)
Sometimes you need multiple plots on the same subplot but with different y-axes.

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

ax2 = ax1.twinx()  # Create a secondary y-axis

ax1.plot(x, y1, 'g-')  # Green sine wave
ax2.plot(x, np.exp(x / 10), 'b-')  # Blue exponential growth

ax1.set_xlabel("X values")
ax1.set_ylabel("Sine Wave", color='g')
ax2.set_ylabel("Exponential Growth", color='b')

plt.show()
```

- Useful when two variables have different scales
- Ensures clarity in multi-variable plots

## Summary of Subplot Methods
| **Method** | **Best For** | **Example** |

| `plt.subplots()` | Best for multiple subplots | `fig, ax = plt.subplots(2,2)` |

| `fig.add_subplot()` | Object-oriented approach | `fig.add_subplot(1,2,1)` |

# Enhancing Matplotlib Plots for a Power BI or Tableau Look

You can make your Matplotlib plots look as professional as those in Power BI or Tableau by using the following styling techniques:

Matplotlib has built-in styles that instantly enhance the appearance of plots.

# Power BI-Style Dashboard Plot Template
This template in Matplotlib and Seaborn includes:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.gridspec as gridspec

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x) / 3
y4 = np.exp(x / 10)

# Set Power BI-like styling
sns.set_theme(style="darkgrid")  # Dark background with grid
palette = sns.color_palette("husl", 4)  # Professional color palette

# Create a dashboard-style layout
fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(2, 2, height_ratios=[3, 1])  # Larger top row

# First subplot: Line chart (Main Trend)
ax1 = plt.subplot(gs[0, :])  # Top row spans across all columns
ax1.plot(x, y1, color=palette[0], linewidth=2, label="Sine Wave")
ax1.plot(x, y2, color=palette[1], linewidth=2, linestyle='dashed', label="Cosine Wave")
ax1.set_title("Main Trend: Sine & Cosine", fontsize=16, fontweight='bold')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.6)

# Annotation
ax1.annotate("Peak Point", xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=12, fontweight='bold')

# Second subplot: Bar Chart
ax2 = plt.subplot(gs[1, 0])  # Bottom-left
ax2.bar(x[::10], y3[::10], color=palette[2])
ax2.set_title("Bar Chart: Scaled Tangent", fontsize=14)
ax2.grid(True, linestyle=':', alpha=0.6)

# Third subplot: Scatter Plot
ax3 = plt.subplot(gs[1, 1])  # Bottom-right
ax3.scatter(x, y4, color=palette[3], alpha=0.7, edgecolors='black')
ax3.set_title("Scatter Plot: Exponential Growth", fontsize=14)
ax3.grid(True, linestyle=':', alpha=0.6)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

- **GridSpec Layout (`gridspec.GridSpec()`)**  Allows complex multi-plot dashboards.

- **Multiple Plot Types:**
  - Line Chart (Main trend)

  - Scatter Plot

```python
import cv2
from google.colab.patches import cv2_imshow  # Import for displaying images in Colab

img_path = f'./PistolData_merged/pistol_images/1.jpg'  # Image path

# Load the image
img = cv2.imread(img_path)  # Reads image in BGR format

# Check if image was loaded properly
if img is None:
    print("Error: Image not found or path is incorrect")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    cv2_imshow(img)  # Use cv2_imshow() instead of cv2.imshow()
```

- `cv2.waitKey(0)` and `cv2.destroyAllWindows()` are also not needed in Google Colab.

- Opens an image in a separate window.
- Reads images in BGR format and converts them to RGB.
- Faster than Matplotlib for large images.
- Not recommended for Jupyter Notebooks as it requires a local environment.

## 4. Using PIL (Pillow)

```python
from PIL import Image

# Load the image using OpenCV
img_path = "./PistolData_merged/pistol_images/1.jpg"
img = cv2.imread(img_path)  # OpenCV loads image as a NumPy array

# Convert NumPy array (BGR) to PIL Image (RGB)
if img is None:
    print("Error: Image not found or path is incorrect")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    img_pil = Image.fromarray(img)  # Convert NumPy array to PIL Image
    img_pil.show()  # Opens in default image viewer
```

## 5. Using Seaborn (`sns.heatmap()`)

```python
import cv2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the image (OpenCV loads it as BGR)
img_path = "./PistolData_merged/pistol_images/1.jpg"
img = cv2.imread(img_path)

# Check if image exists
if img is None:
    print("Error: Image not found or path is incorrect")
else:
    # Convert to grayscale (2D array)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Plot the grayscale image using Seaborn Heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(img_gray, cmap="gray", cbar=False)  # Disable color bar
    plt.title("Seaborn Heatmap Representation of Image")
    plt.show()
```

## 7. Using PyTorch (`torchvision.utils.make_grid`)

```python
import torch
import torchvision
import cv2
import matplotlib.pyplot as plt

# Load the image (OpenCV loads as BGR NumPy array)
img_path = "./PistolData_merged/pistol_images/1.jpg"
img = cv2.imread(img_path)

# Check if image exists
if img is None:
    print("Error: Image not found or path is incorrect")
else:
    # Convert BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Convert NumPy array (H, W, C)  PyTorch tensor (C, H, W)
    img_tensor = torch.from_numpy(img).permute(2, 0, 1).float() / 255.0  # Normalize to [0,1]

    # Create grid (expecting a batch, so we add a batch dimension)
    grid_img = torchvision.utils.make_grid([img_tensor], nrow=1)

    # Convert PyTorch tensor (C, H, W)  NumPy (H, W, C) for Matplotlib
    plt.imshow(grid_img.permute(1, 2, 0))
    plt.axis("off")
    plt.show()
```

- Used for batch image visualization in deep learning.
- Displays multiple images in a grid format.
